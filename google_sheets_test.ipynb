```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

def update_google_sheet():
    # Set up credentials
    scope = [
        'https://spreadsheets.google.com/feeds',
        'https://www.googleapis.com/auth/drive'
    ]
    
    # Path to your service account credentials JSON file
    creds = ServiceAccountCredentials.from_json_keyfile_name(key_file, scope)
    
    # Authorize the client
    client = gspread.authorize(creds)
    
    # Open the Google Sheet
    # You can open by spreadsheet URL or by spreadsheet name
    # Method 1: Open by URL
    # spreadsheet = client.open_by_url('YOUR_SPREADSHEET_URL')
    
    # Method 2: Open by name
    spreadsheet = client.open('linkedin_jobs')
    
    # Select a specific worksheet
    worksheet = spreadsheet.worksheet('linkedin_jobs.csv')  # or spreadsheet.get_worksheet(0)
    
    # Update a specific cell
    worksheet.update_cell(1, 1, 'Stephen Test')
    
    # Update multiple cells
    worksheet.update('A1:B2', [
        ['Value 1', 'Value 2'],
        ['Value 3', 'Value 4']
    ])
    
    # Append a new row
    worksheet.append_row(['New', 'Row', 'Data'])
    
    print("Sheet updated successfully!")

# Example usage
if __name__ == '__main__':
    update_google_sheet()
    
```

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

def manage_google_sheets():
    # Set up credentials
    scope = [
        'https://spreadsheets.google.com/feeds',
        'https://www.googleapis.com/auth/drive'
    ]
    
    # Path to your service account credentials JSON file
    creds = ServiceAccountCredentials.from_json_keyfile_name(key_file, scope)
    
    # Authorize the client
    client = gspread.authorize(creds)
    
    # List all spreadsheets accessible by this service account
    print("Listing all accessible spreadsheets:")
    all_spreadsheets = client.openall()
    for sheet in all_spreadsheets:
        print(f"Spreadsheet Name: {sheet.title}")
    
    # If you want to list sheets within a specific spreadsheet
    try:
        spreadsheet = client.open('Your Spreadsheet Name')
        print("\nSheets within this spreadsheet:")
        for worksheet in spreadsheet.worksheets():
            print(f"Worksheet Name: {worksheet.title}")
    except Exception as e:
        print(f"Error accessing spreadsheet: {e}")
    
    # Print service account email (who owns/can access these sheets)
    print("\nService Account Details:")
    print(f"Service Account Email: {creds.service_account_email}")

# Example usage
if __name__ == '__main__':
    manage_google_sheets()
```
